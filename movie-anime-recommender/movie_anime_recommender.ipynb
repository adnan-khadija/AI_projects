la simularité


```python
def similarite_cosinus(vecteur_a, vecteur_b):
  produit_scalaire= np.dot(vecteur_a,vecteur_b)
  norme_a=np.linalg.norm(vecteur_a)
  norme_b=np.linalg.norm(vecteur_b)
  return produit_scalaire /(norme_a*norme_b)

vec_one_piece=contenus_bruts[0]['vecteur']
vec_naroto=contenus_bruts[3]['vecteur']
resultat= similarite_cosinus(vec_one_piece,vec_naroto)
print('similarité  one peice / naroto:', resultat)
```

```python
def trouver_similarites(index_contenu,contenus_bruts):
  vecteur_cible= contenus_bruts[index_contenu]['vecteur']
  resultats=[]

  for i, contenu in enumerate(contenus_bruts):
    if i==index_contenu:
      continue
    score=similarite_cosinus(vecteur_cible, contenu["vecteur"])
    resultats.append((contenu["titre"],score))
  return resultats
similarites_one_piece=trouver_similarites(0,contenus_bruts)
for titre , score in similarites_one_piece:
  print(titre, ":", round(score,3))
```

Top-5


```python
def top_n_recommandations(index_contenu,contenus_bruts,n=5):
  similarites=trouver_similarites(index_contenu,contenus_bruts)
  similarites_triees=sorted(similarites, key=lambda x:x[1],reverse=True)
  return similarites_triees[:n]
top5_one_piece=top_n_recommandations(0,contenus_bruts)
for titre, score in top5_one_piece:
  print(titre,":",round(score,3))
```

Recommandation par titre

```python
def recommander(titre_recherche, contenus_bruts, n=5):
  index_trouve=None
  for i, contenu in enumerate(contenus_bruts):
      if contenu["titre"].lower()==titre_recherche.lower():
        index_trouve=i
        break
  if index_trouve is None:
    print(f"Désolé, '{titre_recherche}'n'est pas dans le dateset .")
    return
  print(f"Recommandations pour'{contenus_bruts[index_trouve]['titre']}':\n")
  top5=top_n_recommandations(index_trouve, contenus_bruts, n)
  for titre, score in top5:
    print(f"{titre} (score: {round(score,3)})")

recommander('Titre Qui Existe Pas', contenus_bruts)
```
